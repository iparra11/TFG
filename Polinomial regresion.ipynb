```python
def values(dic,investment_id):
    v=[]
    lenth=len(dic[investment_id])
    for i in range(lenth):
        j=int(i)
        if i==0:
            v.append(dic[investment_id][j])
        else:
            v.append(v[-1]+dic[investment_id][j])
    return(v)
```

```python
import pandas as pd
import time

t_0=time.time()
df = pd.read_csv (r'train.csv')
d_train={}
for i in range(1000):
    if df.loc[i,'investment_id'] not in d_train:
        d_train[df.loc[i,'investment_id']]=[]
    for j in range(300):
        j=str(j)
        d_train[df.loc[i,'investment_id']].append(df.loc[i,'f_'+j])
t_1=time.time()
print(f'Time : {t_1-t_0}')
```

```python
def muestras(list,n):
    mues=[]
    m=len(list)
    for i in range(m-n):
        l=[]
        for j in range(n):
            l.append(list[j+i])
        l.append(list[i+j+1])
        mues.append(l)
    return(mues)
```
